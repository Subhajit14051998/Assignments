1: Matplotlib is a popular plotting library for Python. It provides a
wide range of functionalities for creating various types of plots and
visualizations.

Matplotlib is used to generate static, animated, and interactive
visualizations in Python. It is widely used in data analysis, scientific
research, and data visualization tasks.

Five plots that can be plotted using the Pyplot module of Matplotlib
are:

a\) Line plot: A plot where data points are connected by straight lines.

b\) Scatter plot: A plot where individual data points are represented as
markers on a coordinate system.

c\) Bar plot: A plot that represents categorical data with rectangular
bars.

d\) Histogram: A plot that displays the distribution of a continuous
variable by dividing it into bins.

e\) Pie chart: A circular plot that represents the proportion of
different categories in a dataset.

2:A scatter plot is a type of plot or graph that represents the
relationship between two variables. It displays individual data points
as markers on a two-dimensional coordinate system, with one variable
plotted along the x-axis and the other variable plotted along the
y-axis.

To plot a scatter plot using the provided data, we can use Matplotlib's
Pyplot module.

Here's the code:

import matplotlib.pyplot as plt

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)

plt.title("Scatter Plot")

plt.xlabel("X")

plt.ylabel("Y")

plt.show()

This code generates random data for x and y using a normal distribution.
Then it creates a scatter plot using plt.scatter(x, y). Finally, it adds
a title, xlabel, and ylabel to the plot using plt.title(), plt.xlabel(),
and plt.ylabel(), respectively.

3.The subplot() function in Matplotlib is used to create multiple
subplots within a single figure. It allows you to arrange multiple plots
in a grid-like structure. Each subplot can display different data or a
different representation of the same data.

To draw four line plots using the subplot() function, we can use the
provided data and the following code:

import matplotlib.pyplot as plt

import numpy as np

x = np.array(\[0, 1, 2, 3, 4, 5\])

\# Line 1

y1 = np.array(\[0, 100, 200, 300, 400, 500\])

plt.subplot(2, 2, 1)

plt.plot(x, y1)

plt.title("Line 1")

\# Line 2

y2 = np.array(\[50, 20, 40, 20, 60, 70\])

plt.subplot(2, 2, 2)

plt.plot(x, y2)

plt.title("Line 2")

\# Line 3

y3 = np.array(\[10, 20, 30, 40, 50, 60\])

plt.subplot(2, 2, 3)

plt.plot(x, y3)

plt.title("Line 3")

\# Line 4

y4 = np.array(\[200, 350, 250, 550, 450, 150\])

plt.subplot(2, 2, 4)

plt.plot(x, y4)

plt.title("Line 4")

plt.tight_layout()

plt.show()

4.A bar plot, also known as a bar chart, is a visualization that
represents categorical data with rectangular bars. The length of each
bar corresponds to the value of the category it represents.

Bar plots are used to compare and display the values of different
categories or groups.

To plot a bar plot and a horizontal bar plot using the provided data,we
can use the following code:

import matplotlib.pyplot as plt

import numpy as np

company = np.array(\["Apple", "Microsoft", "Google", "AMD"\])

profit = np.array(\[3000, 8000, 1000, 10000\])

plt.figure(figsize=(8, 6))

\# Bar plot

plt.subplot(1, 2, 1)

plt.bar(company, profit)

plt.title("Bar Plot")

plt.xlabel("Company")

plt.ylabel("Profit")

\# Horizontal bar plot

plt.subplot(1, 2, 2)

plt.barh(company, profit)

plt.title("Horizontal Bar Plot")

plt.xlabel("Profit")

plt.ylabel("Company")

plt.tight_layout()

plt.show()

5: A box plot, also known as a whisker plot, is a graphical
representation of the distribution of numerical data through quartiles.
It displays the minimum, first quartile, median, third quartile, and
maximum values of a dataset, as well as any potential outliers.

Box plots are useful for detecting outliers, comparing distributions,
and visualizing summary statistics.

To plot a box plot using the provided data, we can use the following
code:

import matplotlib.pyplot as plt

import numpy as np

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

data = \[box1, box2\]

plt.boxplot(data)

plt.title("Box Plot")

plt.xlabel("Box")

plt.ylabel("Value")

plt.show()
