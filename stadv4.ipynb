1.A t-test is used when the sample size is small, and the population
standard deviation is unknown. A z-test is used when the sample size is
large, and the population standard deviation is known. For example, if
you wanted to test if a new drug reduces blood pressure, you might use a
t-test with a small sample size and unknown variance. If you were
comparing the heights of a large group of adults to a known population
mean, you might use a z-test.

2.In a one-tailed test, you’re interested in only one direction of the
effect (e.g., greater than or less than). In a two-tailed test, you’re
interested in both directions (e.g., not equal to). For instance, if
you’re testing whether a new fertilizer increases crop yield, a
one-tailed test would focus on whether the yield is greater, while a
two-tailed test would consider if the yield is significantly different
in either direction.

3.Type 1 error (False Positive) occurs when you incorrectly reject a
true null hypothesis. Type 2 error (False Negative) occurs when you fail
to reject a false null hypothesis. Example of Type 1 error: A medical
test wrongly indicating a healthy person as having a disease. Example of
Type 2 error: Failing to detect a faulty product during quality control.

4.Bayes’s theorem is a way to update the probability of a hypothesis
based on new evidence. It involves the prior probability, likelihood,
and marginal likelihood. For example, consider a medical test for a rare
disease: You use Bayes’s theorem to update the likelihood of having the
disease based on the test result and the disease prevalence.

5.A confidence interval is a range around a sample statistic (like a
mean) that likely contains the true population parameter. It quantifies
the uncertainty of our estimate. To calculate a confidence interval, you
determine the margin of error using the standard error and a chosen
level of confidence (e.g., 95%). For example, if you calculated a mean
height of 170 cm with a margin of error of 2 cm and a confidence level
of 95%, the confidence interval would be 168 cm to 172 cm.

7.The 95% confidence interval for a sample mean of 50 with a standard
deviation of 5 is approximately 50 ± 1.96 \* (5 / √n), where n is the
sample size. Interpreted, this means you’re 95% confident that the true
population mean lies between 48.02 and 51.98.

8.The margin of error in a confidence interval represents the range
within which the true population parameter is likely to lie. A larger
sample size reduces the margin of error, as it increases the precision
of the estimate. For example, a survey with a sample size of 100 might
have a margin of error of 5%, while a survey with a sample size of 1000
might have a margin of error of 1.6%.
