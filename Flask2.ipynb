1\. GET and POST methods are HTTP methods used in web development to
send and receive data between a client (such as a web browser) and a
server.

a\) GET: This method is used to request data from a specified resource.
It retrieves data from the server and includes the data in the URL as
query parameters. It is a safe and idempotent method, meaning it should
not have any side effects on the server and can be called multiple times
without changing the server's state.

b\) POST: This method is used to submit data to be processed to a
specified resource. It sends data in the body of the HTTP request rather
than in the URL. POST requests are not idempotent, meaning they can have
side effects on the server, such as creating or updating data.

2\. In Flask, the request object is used to access incoming request
data. It allows you to retrieve data sent by a client in the form of
HTTP parameters, form data, JSON, files, etc. The request object
provides methods and attributes to access different parts of the
request, such as headers, cookies, query parameters, and the request
body.

By using the request object, we can extract and process the data
submitted by the client, which is essential for building dynamic web
applications that respond to user input.

3\. The redirect() function in Flask is used to redirect the client's
web browser to a different location. It sends a response to the client
with a status code of 302 (Found) or 301 (Moved Permanently) and sets
the 'Location' header to the provided URL.

Redirects are commonly used to redirect users after a successful form
submission, after authentication, or to redirect from one route to
another.

4\. Templates in Flask are files that contain HTML code with
placeholders for dynamic content. They allow you to separate the
presentation logic from the application logic. Templates provide a way
to generate dynamic HTML pages by inserting values from variables or
data retrieved from the server.

The render_template() function in Flask is used to render a template by
providing the template filename and the data to be passed to the
template. It combines the template with the provided data to generate
the final HTML output, which can then be sent as a response to the
client.
