1: A t-test is used when the sample size is small or when the population
standard deviation is unknown, while a z-test is used when the sample
size is large and the population standard deviation is known. For
example, you might use a t-test to compare the means of two small
groups, such as the effectiveness of two different teaching methods on a
class of 20 students. On the other hand, a z-test could be used to
compare the heights of a large population of adults to a known average
height.

2: In a one-tailed test, you’re testing for a specific direction of
difference or effect (e.g., greater than or less than), while in a
two-tailed test, you’re testing for any significant difference or effect
(either greater or less than). For instance, in a one-tailed test, you
might investigate whether a new drug increases reaction times, while a
two-tailed test could determine if the new drug has any effect on
reaction times at all.

3: Type 1 error (False Positive) occurs when you mistakenly reject a
true null hypothesis, while Type 2 error (False Negative) happens when
you fail to reject a false null hypothesis. An example of Type 1 error
could be a medical test falsely diagnosing a healthy person as having a
disease. An example of Type 2 error might be a medical test failing to
detect a disease that a person actually has.

4: Bayes’s theorem is a mathematical formula that calculates the
probability of a hypothesis being true given new evidence. It’s used to
update our beliefs based on new information. For example, if you want to
determine the probability that a person has a certain disease given a
positive test result, you’d use Bayes’s theorem to factor in the
prevalence of the disease in the population and the accuracy of the
test.

5: A confidence interval is a range of values within which a population
parameter, like a mean or proportion, is likely to fall with a certain
level of confidence. To calculate a confidence interval, you use sample
data and a margin of error. For instance, if you measured the heights of
a sample of 100 people and calculated a mean height of 165 cm with a 95%
confidence interval of ±3 cm, you can be 95% confident that the true
mean height of the entire population falls within the range of 162 cm to
168 cm.

    maximum likely difference between the sample statistic (such as the
    sample mean) and the true population parameter. A larger sample size
    generally results in a smaller margin of error.

sample size, like 100 people, the margin of error might be ±8%. However,
with a larger sample size of 1000 people, the margin of error could
decrease to ±3%. This means you can be more confident that the sample
mean is closer to the true population mean.
